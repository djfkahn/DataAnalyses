## Kernel Density Estimators or KDEs...

In constructing a histogram, we need to consider the width of the bins (dividing a column of data into equal sub-intervals) and the end points of each bin (where each bin starts and ends). So the shape of the distribution can vary based on how many bins we choose and the size of each individual bin. And that's not good. KDEs alleviate these problems.

KDEs belong to a class of estimators called the non-parametric probability density estimators. A typical parametric estimator on the other hand has a fixed functional form (structure) and the parameters of that function are the only things we need to recreate and characterize that distribution. Non-parametric estimators (KDEs) have no fixed structure and hence depend upon all the data points to reach an estimate of the distribution.

```python
sns.distplot(df['Loan Size'], color = 'red')
plt.title('Loan Size Distribution', fontsize=18)
plt.xlabel('Loan Size', fontsize=16)
plt.ylabel('Density', fontsize=16)
```

Without histogram...

```python
sns.distplot(df['Loan Size'], color = 'red', hist = False)
plt.title('Loan Size Distribution', fontsize=18)
plt.xlabel('Loan Size', fontsize=16)
plt.ylabel('Density', fontsize=16)
```

## Discretization

We can turn a numeric variable into an ordinal one through one of two functions.

- For <b>equal-width binning</b>, use <b>pandas.cut</b>. All bins will be of the same width.
- For <b>equal-frequency binning</b>, use <b>pandas.qcut</b>. All bins will (attempt to) have the same number of records.

<b>Example</b>: Discretizing Loan Size into 5 bins of equal width.

```python
df2 = df.copy()
df2['Discretized_Loan_Size'] = pd.cut(df2['Loan Size'], 5)
```

```python
df2.Discretized_Loan_Size.unique()
```

```python
sns.countplot(df2['Discretized_Loan_Size'])
plt.title('Count by Loan Size', fontsize=18)
plt.xlabel('Discretized Loan Size', fontsize=16)
plt.ylabel('Count', fontsize=16)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(df2['Discretized_Loan_Size'])
plt.title('Count by Loan Size', fontsize=18)
plt.xlabel('Discretized Loan Size', fontsize=16)
plt.ylabel('Count', fontsize=16)
```

<b>Example</b>: Discretizing Loan Size into 5 bins of equal frequency.

```python
df2 = df.copy()
df2['Discretized_Loan_Size'] = pd.qcut(df2['Loan Size'], 5)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(df2['Discretized_Loan_Size'])
plt.title('Count by Loan Size', fontsize=18)
plt.xlabel('Discretized Loan Size', fontsize=16)
plt.ylabel('Count', fontsize=16)
```

## One Categorical, One Numeric...

A box and whisker plot is one of the simplest ways to visualize the underlying structure of the data. The horizontal line somewhere in the box is the median and the two ends of the box represent the 25th and 75th percentile of the data. Median of course is the center point of a sorted column of data. Half the data points are above that median and the remaining half are below. From the 25th percentile (lower edge of the box), 75% of the data points are above that mark and 25% are below that. And for the 75th percentile, 25% of the data points are above that mark and 75% of them are below.

The two whiskers that stick out at the top and the bottom of the box are the first 25% and the last 25% of the datapoints. The points beyond the two ends of the whiskers are outliers.

A sample box plot shown below displays the <i>Loan Size</i> by <i>Loan Grade</i>.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x = 'Loan Grade', y = 'Loan Size', data = df, order = ['AAA', 'AA', 'A', 'B', 'C', 'D', 'E'])
plt.title('Loan Size by Loan Grade', fontsize=18)
plt.xlabel('Loan Grade', fontsize=16)
plt.ylabel('Loan Size', fontsize=16)
```

So a box plot is convenient for comparing summary statistics such as range and quartiles but it doesn't let you see variations in the data. If the distribution is multimodal (with multiple peaks), a box plot is not going to reveal that. And that can be a problem sometimes.

We used a box plot to depict the median and the quartiles, we can use a bar plot for the mean. Shown below is the mean <b>Loan Size</b> by <b>Loan Grade</b>.

```python
sns.catplot(x = 'Loan Grade', y = 'Loan Size', data = df, kind = 'bar', order = ['AAA', 'AA', 'A', 'B', 'C', 'D', 'E'])
plt.title('Loan Size by Loan Grade', fontsize=18)
plt.xlabel('Loan Grade', fontsize=16)
plt.ylabel('Loan Size', fontsize=16)
```

```python
plt.figure(figsize=(12, 6))
sns.catplot(x = 'Loan Grade', y = 'Loan Size', data = df, kind = 'bar', order = ['AAA', 'AA', 'A', 'B', 'C', 'D', 'E'])
plt.title('Loan Size by Loan Grade', fontsize=18)
plt.xlabel('Loan Grade', fontsize=16)
plt.ylabel('Loan Size', fontsize=16);
```

The reason for this is that the higher level plotting functions in Seaborn (what the documentation calls figure-level interfaces) have a different way of managing size, largely due to the fact that they often produce multiple subplots. To set the size when using catplot(), relplot(), pairplot(), lmplot() or jointplot(), we use the height keyword to control the size and the aspect keyword to control the shape.

```python
sns.catplot(x = 'Loan Grade', y = 'Loan Size', data = df, kind = 'bar', order = ['AAA', 'AA', 'A', 'B', 'C', 'D', 'E'], 
            height = 5, aspect = 3)
plt.title('Loan Size by Loan Grade', fontsize=18)
plt.xlabel('Loan Grade', fontsize=16)
plt.ylabel('Loan Size', fontsize=16)
```

```python
sns.catplot(x = 'Loan Grade', y = 'Loan Size', row = 'Delinquency Status', data = df, kind = 'bar', order = ['AAA', 'AA', 'A', 'B', 'C', 'D', 'E'], 
            height = 5, aspect = 3)
```

## Confidence Intervals...

Those black lines at the top of each bar identify the confidence interval range for the mean of <i>Loan Size</i> by <i>Loan Grade</i>. So what are they? This dataset as with all other future datasets is still a sample from a population of all customers that would ever be on the Upstart platform. So a 99% confidence interval means that 99% of the samples drawn from this population of customers will include the true population's mean <i>Loan Size</i> within that range specified by those black lines and 1% of the samples won't. So there is a 1-in-100 chance that the confidence interval does not include the true population mean of <i>Loan Size</i> by each <i>Loan Grade</i> for a given sample.

Similarly, A 95% confidence interval means that 95% of the samples drawn from a population are likely to include the true population mean and 5% won't. So naturally this is a bit more aggressive and hence those black lines at the top are likely to have a smaller span as shown below than that for a 99% confidence level.

## With 95% confidence level...

```python
sns.catplot(x = 'Loan Grade', y = 'Loan Size', data = df, kind = 'bar', ci = 95, order = ['AAA', 'AA', 'A', 'B', 'C', 'D', 'E'])
plt.title('Loan Size by Loan Grade', fontsize=18)
plt.xlabel('Loan Grade', fontsize=16)
plt.ylabel('Loan Size', fontsize=16);
```

But why does the mean for <i>Loan Grade</i> AAA have a larger confidence interval band than say the mean for <i>Loan Grade</i> B? An attempt at understanding that...

But before that, this is how we estimate confidence intervals... 

$$\Large \bar{x} \pm  z \times \frac{s}{\sqrt{n}}$$

x_bar is the sample mean of <i>Loan Size</i> by <i>Loan Grade</i>, z is the z-score corresponding to the confidence level (z = 1.96 at 95% confidence level and 2.576 at 99% confidence level) and s/sqrt(n) is the standard error associated with the sampling distribution of the sample mean of <i>Loan Size</i> by each <i>Loan Grade</i>. The bigger the sample size by <i>Loan Grade</i>, the smaller the sampling error and the better it gets in trying to capture the true population mean. And that intuitively makes sense. The more instances there are of <i>Loan Size</i> for each <i>Loan Grade</i>, the more precise an estimate of the true population mean <i>Loan Size</i> we will have by each loan grade.

Back to the question at hand, why does the mean for <i>Loan Grade</i> AAA have larger confidence interval than say the mean for <i>Loan Grade</i> B?

```python
df.groupby(by = 'Loan Grade')['Loan Size'].size()
```

That's because there are only 6 instances for <i>Loan Grade AAA</i> and 3008 instances for <i>Loan Grade B</i>. That sqrt(n) in the denominator reduces the standard error associated with estimating the population mean for <i>Loan Grade B</i> and hence a smaller confidence interval.

Seaborn though uses a bootstrap approach at estimating confidence intervals but a similar logic applies.

```python
sns.catplot(x='Loan Grade', y='APR', data=df, kind='bar', ci=95, order = ['AAA', 'AA', 'A', 'B', 'C', 'D', 'E'])
```

```python
n = df.groupby(by = 'Loan Grade')['APR'].size()
```

```python
x = df.groupby(by = 'Loan Grade')['APR'].mean()
```

```python
s = df.groupby(by = 'Loan Grade')['APR'].std()
```

```python
s
```

```python
x + (1.96 * s / (n**(1/2)))
```

```python
x - (1.96 * s / (n**(1/2)))
```

## Violin Plots...

But sometimes, the median and the mean arenâ€™t enough to understand a dataset. Are most of the values clustered around the median? Or are they clustered around the minimum and the maximum with nothing in the middle? When we have questions like these, distribution plots are there to help and that is where violin plots come in.

A violin plot is a hybrid of a box plot and a kernel density plot, displaying quartiles as well as peaks and valleys in the data.

```python
sns.violinplot(x = 'Loan Grade', y = 'Loan Size', data = df, order = ['AAA', 'AA', 'A', 'B', 'C', 'D', 'E'])
```

```python
sns.violinplot(x = 'Loan Grade', y = 'APR', data=df, order = ['AAA', 'AA', 'A', 'B', 'C', 'D', 'E'])
```

So we see a big difference in distribution of <i>APR</i> for <i>Loan Grade AAA</i> vs. say <i>Loan Grade B</i>.

And this is what a box plot would have revealed for the same data. 

```python
sns.catplot(x = 'Loan Grade', y = 'APR', data=df, kind='box', order = ['AAA', 'AA', 'A', 'B', 'C', 'D', 'E'])
```

Not as rich in information.
